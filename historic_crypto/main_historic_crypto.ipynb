```python
import pandas as pd
import matplotlib.pyplot as plt

from historic_crypto import *
from backtest import *
from strategy_1 import *
```

```python
pairs = get_pairs()
```

```python
pairs_pnl, total_pnl = backtest(pairs=pairs, end_date='now', nb_days=365, amount=10)
```

```python
pairs_pnl = []
horizon = [365, 365*2, 365*3, 365*4, 365*5]

for nb_days in horizon:
    print(f'Backtesting {nb_days} Days ----------')
    _, tot_pnl = backtest(pairs=pairs, end_date='now', nb_days=nb_days, amount=10, verbose=False)
    pairs_pnl.append(tot_pnl)
    print(f'---------- PnL = {tot_pnl} USDT \n')
```

```python
plt.plot(horizon, pairs_pnl)
plt.xlabel('Horizon - Days')
plt.ylabel('Total PnL - USDT')
plt.title('Total PnL vs Horizon')
plt.show()
```
